```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Criando o diagrama
fig, ax = plt.subplots(figsize=(12, 8))

# Adicionando blocos e setas
# Embedding pr-treinado
ax.text(0.2, 0.9, "Corpus Grande (Ex.: Wikipedia)", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="lightblue"))
ax.arrow(0.2, 0.85, 0, -0.1, head_width=0.02, head_length=0.05, fc='black', ec='black')
ax.text(0.2, 0.7, "FastText / Word2Vec / GloVe", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="lightgreen"))
ax.arrow(0.2, 0.65, 0, -0.1, head_width=0.02, head_length=0.05, fc='black', ec='black')
ax.text(0.2, 0.5, "Embedding Pr-treinado (Matriz de Vetores)", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="lightcoral"))
ax.arrow(0.2, 0.45, 0.2, -0.15, head_width=0.02, head_length=0.05, fc='black', ec='black')

# Embedding no modelo
ax.text(0.6, 0.9, "Corpus de Treinamento (Pequeno)", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="lightblue"))
ax.arrow(0.6, 0.85, 0, -0.1, head_width=0.02, head_length=0.05, fc='black', ec='black')
ax.text(0.6, 0.7, "Tokenizao (TensorFlow)", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="lightgreen"))
ax.arrow(0.6, 0.65, 0, -0.1, head_width=0.02, head_length=0.05, fc='black', ec='black')
ax.text(0.6, 0.5, "Tokens (ndices Numricos)", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="lightcoral"))
ax.arrow(0.6, 0.45, -0.2, -0.15, head_width=0.02, head_length=0.05, fc='black', ec='black')

# Juno dos Embeddings
ax.text(0.4, 0.25, "Camada de Embedding", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="gold"))
ax.arrow(0.4, 0.2, 0, -0.1, head_width=0.02, head_length=0.05, fc='black', ec='black')
ax.text(0.4, 0.05, "Modelo Neural (Ex.: LSTM, GRU)", fontsize=12, ha="center", bbox=dict(boxstyle="round", facecolor="salmon"))

# Conexes
ax.arrow(0.2, 0.25, 0.2, 0, head_width=0.02, head_length=0.02, fc='black', ec='black')
ax.arrow(0.6, 0.25, -0.2, 0, head_width=0.02, head_length=0.02, fc='black', ec='black')

# Ajustando o layout
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.axis('off')

# Ttulo
plt.title("Diagrama de Embeddings Pr-treinados e Usados em Arquitetura", fontsize=14)
plt.show()
```
